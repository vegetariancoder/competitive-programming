```python
from pyspark.sql.session import SparkSession
from pyspark.sql.functions import col, min, year,count

spark = SparkSession \
    .builder \
    .config("spark.jars",
            "/Users/sahilnagpal/Downloads/mysql-connector-java-8.0.29.jar") \
    .master("local[*]") \
    .appName("leetcode-problems") \
    .getOrCreate()

business_operations = spark.read.format("jdbc"). \
    option("url", "jdbc:mysql://localhost:3306/namastesql") \
    .option("driver", "com.mysql.cj.jdbc.Driver") \
    .option("dbtable", "business_operations") \
    .option("user", "root") \
    .option("password", "password") \
    .load()
```

```python
business_operations.show()
```

```python
business_operations\
    .groupBy(col("city_id"))\
    .agg(min(year(col("business_date"))).alias("first_year"))\
    .groupBy(col("first_year").alias("first_operation_year"))\
    .agg(count("city_id").alias("no_of_new_cities"))\
    .orderBy(col("first_operation_year"))\
    .show()
```
